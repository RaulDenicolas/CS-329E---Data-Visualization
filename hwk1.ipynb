**Q. 5 (10 points)**

**Q. 9 (10 points)**.  
The Scandinavian countries are Denmark, Norway, and Sweden. For each of the years (from 1980 to 2013) find the difference between the country with the highest immigration and the country with the lowest immigration. Print the year and the difference.

```python
scandinvian = ['Denmark', 'Norway', 'Sweden']
df_scandinvian = df.loc[scandinvian]
#print(df_scandinvian)
df_scandinvian = df_scandinvian.iloc[:, 2:-1]
#print(df_scandinvian)
for column in df_scandinvian.columns:
    max_immigration = df_scandinvian[column].max()
    min_immigration = df_scandinvian[column].min()
    difference = max_immigration - min_immigration
    print(column, difference)
```
